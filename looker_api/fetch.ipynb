```python
import looker_sdk
from looker_sdk.sdk.api40.methods import Looker40SDK
import pandas as pd 
from io import StringIO
```

```python
sdk: Looker40SDK = looker_sdk.init40()

response = sdk.run_look(
    look_id="7",

    result_format="csv")
```

```python
#dashboard usage

response_file = StringIO(response)

df = pd.read_csv(response_file)

df.head(15)

```

```python
import pandas_gbq

# Set your project ID and dataset ID
project_id = 'astrafy-sandbox-greta'
dataset_id = 'system_monitoring'

# Set the table name
table_name = 'Looker_Query_usage'

# Push the dataframe to BigQuery
pandas_gbq.to_gbq(df, f'{project_id}.{dataset_id}.{table_name}', project_id=project_id, if_exists='replace')
```
